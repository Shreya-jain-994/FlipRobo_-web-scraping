```python
for i in Product_url:
    driver.get(i)
    time.sleep(3)
    
    
            # Fetching brand name
    try:
        brand= driver.find_element_by_xpath('//a[@id="bylineInfo"]')
        Brand.append(brand.text.replace("Brand :"," "))
    except  NoSuchElementException:
            Brand.append('-')
            
            
        # Fetching Product name
    try:
        product= driver.find_element_by_xpath('//span[@id="productTitle"]')
        Product.append(product.text)
    except  NoSuchElementException:
            Product.append('-')
            
        #fetching rating details    

    try:
        rating_=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']").click()
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Rating.append(rating.text)
    except NoSuchElementException as e:
        Rating.append('-')
        

    #fetching price details    
    try:
        price=driver.find_element_by_xpath('//*[@id="priceblock_ourprice"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
            
    #fetching exchange details    
    try:
        Exchange=driver.find_element_by_xpath('//*[@id="icon-farm-container"]/div/div[1]/span/div[2]/a')
        Return_Exchange.append(Exchange.text)
    except NoSuchElementException:
        Return_Exchange.append('-')
        

        # Fetching Expected_Delivery details 
    try:
        Delivery=driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]/b')
        Expected_Delivery.append(Delivery.text)
    except  NoSuchElementException:
            Expected_Delivery.append('-')
            
            
    #fetching availability details    
    try:
        availability=driver.find_element_by_xpath('//*[@id="availability"]/span')
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
        
    #fetching other details    
    try:
        details=driver.find_element_by_xpath('//*[@id="feature-bullets"]/ul')
        Other_details.append(details.text.replace("\n"," "))
    except NoSuchElementException:
        Other_details.append('-')
        
        
        #fetching no of rating details 
    No_of_rating=[]
    try:
        ratings=driver.find_element_by_xpath('//div[@id="averageCustomerReviews"]')
        No_of_rating.append(ratings.text)
    except NoSuchElementException:
        No_of_rating.append('-')
        
```

```python
# Fetching name details
name=driver.find_elements_by_xpath('//h2[@class="title title-6"]')
for i in name:
    try:
        hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
# Fetching distance details

distance=driver.find_elements_by_xpath('//span[@class="description"]')
for i in distance:
    try:
        Distance.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
        Distance.append('-')

# Fetching price details

cost=driver.find_elements_by_xpath('//div[@class="price title-5"]')
for i in cost:
    try:
        Price.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
            Price.append('-')
            
# Fetching rating details            
rating=driver.find_elements_by_xpath('//div[@class="score orange big"]')
for i in rating:
    try:
        Rating.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
        Rating.append('-')
```

```python
#creating the dataset
hostel=pd.DataFrame({})
hostel['Name']=hostel_name[:10]
hostel['Distance']=Distance[:10]
hostel['Rating']=Rating[:10]
hostel['Price']=Price[:10]
hostel
```
